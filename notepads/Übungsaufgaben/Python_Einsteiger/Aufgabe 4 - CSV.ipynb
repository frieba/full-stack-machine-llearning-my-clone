```python
import pandas as pd

# 1) CSV laden
df = pd.read_csv("sales.csv")

print("Originaler DataFrame:")
print(df)
print()
```

```python
# 2) Neue Spalte 'umsatz'
df["umsatz"] = df["menge"] * df["preis"]

print("Mit Umsatzspalte:")
print(df)
print()
```

```python
# 3) Gesamtumsatz pro Produkt
umsatz_pro_produkt = df.groupby("produkt")["umsatz"].sum().reset_index()

print("Gesamtumsatz pro Produkt:")
print(umsatz_pro_produkt)
print()
```

```python
# 4) Filter: Umsatz > 3.0
df_hoch = df[df["umsatz"] > 3.0]

print("Zeilen mit Umsatz > 3.0:")
print(df_hoch)
print()
```

```python
# 5) Durchschnittlicher Preis pro Tag
durchschnitt_preis_pro_tag = (
    df.groupby("datum")["preis"].mean().reset_index()
)

print("Durchschnittlicher Preis pro Tag:")
print(durchschnitt_preis_pro_tag)
print()
```

```python
# 6) DataFrame nach Umsatz sortieren (absteigend)
df_sortiert = df.sort_values(by="umsatz", ascending=False)

print("Nach Umsatz sortiert:")
print(df_sortiert)
```
