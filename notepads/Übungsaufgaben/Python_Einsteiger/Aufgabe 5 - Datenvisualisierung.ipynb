```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten
df = pd.DataFrame({
    "sensor": ["A", "A", "A", "B", "B", "B", "C", "C", "C"],
    "wert":   [4.2, 5.1, 4.9, 7.0, 6.8, 7.2, 3.5, 3.7, 3.8],
    "zeit":   [1, 2, 3, 1, 2, 3, 1, 2, 3]
})
```

```python
# --- 1) Histogramm ---
plt.figure(figsize=(6, 4))
plt.hist(df["wert"], bins=5, edgecolor="black")
plt.title("Histogramm der Messwerte")
plt.xlabel("Wert")
plt.ylabel("Häufigkeit")
plt.show()
```

```python
# --- 2) Scatterplot ---
plt.figure(figsize=(6, 4))
plt.scatter(df["zeit"], df["wert"])
plt.title("Scatterplot: Zeit vs. Wert")
plt.xlabel("Zeit")
plt.ylabel("Wert")
plt.show()
```

```python
# --- 3) Barplot: Durchschnitt pro Sensor ---
mittelwerte = df.groupby("sensor")["wert"].mean()

plt.figure(figsize=(6, 4))
plt.bar(mittelwerte.index, mittelwerte.values, color=["red", "green", "blue"])
plt.title("Durchschnittlicher Wert pro Sensor")
plt.xlabel("Sensor")
plt.ylabel("Durchschnittlicher Wert")
plt.show()
```

```python
# --- 4 & Bonus) Scatterplot gefärbt pro Sensor ---
farben = {"A": "red", "B": "green", "C": "blue"}
farben_liste = df["sensor"].map(farben)

plt.figure(figsize=(6, 4))
plt.scatter(df["zeit"], df["wert"], c=farben_liste)
plt.title("Zeit vs. Wert (nach Sensor gefärbt)")
plt.xlabel("Zeit")
plt.ylabel("Wert")
plt.show()
```
