# Matrix Faktorisierung

## Imports

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import NMF
```

## Datengrundlage

```python
data = pd.DataFrame(np.random.randint(0,2,size=(1000,1000))*np.random.randint(0,4,size=(1000,1000)),columns=["Film {}".format(i+1) for i in range(1000)],index=["User {}".format(i+1) for i in range(1000)])
```

```python
data
```

## Matrix Faktorisierung

```python
model = NMF(n_components=5,max_iter=1000)
user_vectors = model.fit_transform(data)
item_vectors = model.components_
```

```python
np.dot(user_vectors,item_vectors)
```
