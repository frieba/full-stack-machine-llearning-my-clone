```python
# Ausgangsdaten
messungen = [
    {"stadt": "Berlin",   "temperatur": 18, "bewoelkt": True},
    {"stadt": "Hamburg",  "temperatur": 16, "bewoelkt": False},
    {"stadt": "München",  "temperatur": 21, "bewoelkt": True},
    {"stadt": "Köln",     "temperatur": 19, "bewoelkt": False},
    {"stadt": "Dresden",  "temperatur": 22, "bewoelkt": True},
]
```

```python
# 1) Durchschnittliche Temperatur berechnen
summe_temp = 0
for eintrag in messungen:
    summe_temp += eintrag["temperatur"]

durchschnitt = summe_temp / len(messungen)
print("Durchschnittstemperatur:", durchschnitt)
```

```python
# 2) Liste der Städte mit Bewölkung
bewoelkte_staedte = []
for eintrag in messungen:
    if eintrag["bewoelkt"]:
        bewoelkte_staedte.append(eintrag["stadt"])

print("Bewölkte Städte:", bewoelkte_staedte)
```

```python
# 3) Temperaturen über dem Durchschnitt
temperaturen_ueber_durchschnitt = []
for eintrag in messungen:
    if eintrag["temperatur"] > durchschnitt:
        temperaturen_ueber_durchschnitt.append(eintrag["temperatur"])

print("Temperaturen über dem Durchschnitt:", temperaturen_ueber_durchschnitt)
```

```python
# 4) Neue Liste mit zusätzlicher Kategorie ('warm' oder 'kühl')
messungen_erweitert = []
for eintrag in messungen:
    neu = eintrag.copy()

    if neu["temperatur"] >= 20:
        neu["kategorie"] = "warm"
    else:
        neu["kategorie"] = "kühl"

    messungen_erweitert.append(neu)

print("Erweiterte Messungen:")
for m in messungen_erweitert:
    print(m)
```
