```python
# Ausgangsdaten
sensordaten = [
    {"sensor": "A", "werte": [4.2, 5.1, 4.9]},
    {"sensor": "B", "werte": [7.0, 6.8, 7.2]},
    {"sensor": "C", "werte": [3.5, 3.7, 3.8]},
    {"sensor": "A", "werte": [5.0, 4.8, 5.2]},
]
```

```python
# 1) Mittelwert-Funktion
def mittelwert(liste):
    return sum(liste) / len(liste)
```

```python
# 2) Durchschnittswerte pro Sensor
def durchschnittswerte_pro_sensor(daten):
    messwerte = {}

    # Alle Werte pro Sensor sammeln
    for eintrag in daten:
        sensor = eintrag["sensor"]
        werte = eintrag["werte"]

        if sensor not in messwerte:
            messwerte[sensor] = []
        messwerte[sensor].extend(werte)

    # Durchschnitt für jeden Sensor berechnen
    ergebnis = {}
    for sensor, werte in messwerte.items():
        ergebnis[sensor] = mittelwert(werte)

    return ergebnis
```

```python
# 3) Sensor mit höchstem Durchschnitt
def sensor_mit_hoechstem_durchschnitt(daten):
    durschnittswerte = durchschnittswerte_pro_sensor(daten)
    return max(durschnittswerte, key=durschnittswerte.get)
```

```python
# 4) Nutzung der Funktionen
print("Durchschnittswerte pro Sensor:")
print(durchschnittswerte_pro_sensor(sensordaten))

print("Sensor mit höchstem Durchschnitt:")
print(sensor_mit_hoechstem_durchschnitt(sensordaten))
```
