```python
# Ausgangsdaten
logs = [
    {"user": "Alice",   "aktionen": ["login", "view", "click"]},
    {"user": "Bob",     "aktionen": ["login", "view"]},
    {"user": "Alice",   "aktionen": ["view", "click", "click"]},
    {"user": "Diana",   "aktionen": ["login"]},
    {"user": "Bob",     "aktionen": ["click", "click"]},
]
```

```python
# 1) Gesamte Anzahl aller Aktionen zählen
aktions_counts = {}

for eintrag in logs:
    for aktion in eintrag["aktionen"]:
        if aktion not in aktions_counts:
            aktions_counts[aktion] = 0
        aktions_counts[aktion] += 1

print("Gesamte Aktionen:", aktions_counts)
```

```python
# 2) Aktionen pro Nutzer zählen
aktionen_pro_nutzer = {}

for eintrag in logs:
    user = eintrag["user"]
    anzahl = len(eintrag["aktionen"])

    if user not in aktionen_pro_nutzer:
        aktionen_pro_nutzer[user] = 0
    aktionen_pro_nutzer[user] += anzahl

print("Aktionen pro Nutzer:", aktionen_pro_nutzer)
```

```python
# 3) Nutzer mit den meisten Aktionen finden
top_user = max(aktionen_pro_nutzer, key=aktionen_pro_nutzer.get)
print("Nutzer mit den meisten Aktionen:", top_user)
```

```python
# 4) Anzahl einzigartiger Aktionen pro Nutzer
unique_actions_per_user = {}
```

```python
# Zuerst alle Aktionen pro Nutzer sammeln
aktionen_sets = {}

for eintrag in logs:
    user = eintrag["user"]
    if user not in aktionen_sets:
        aktions_sets[user] = set()
    aktions_sets[user].update(eintrag["aktionen"])

# Dann die Anzahl der einzigartigen Aktionen zählen
for user, aktionen_set in aktions_sets.items():
    unique_actions_per_user[user] = len(aktionen_set)

print("Einzigartige Aktionen pro Nutzer:", unique_actions_per_user)
```
